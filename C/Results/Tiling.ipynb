```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('tiling.csv','r') as f:
    lines = f.readlines()
    
info = {}
for l in lines:
    a = [a.strip() for a in l.split(',')]
    temp = a[1]
    a[1] = a[2]
    a[2] = temp
    if a[0] not in info:
        info[a[0]] = {}
    if a[1] not in info[a[0]]:
        info[a[0]][a[1]] = {}
    if a[2] not in info[a[0]][a[1]]:
        info[a[0]][a[1]][a[2]] = []
    info[a[0]][a[1]][a[2]].append(a[3])
    
for a in info.keys():
    for b in info[a].keys():
        for c in info[a][b].keys():
            print(info[a][b][c])
            info[a][b][c] = np.mean([float(i) for i in info[a][b][c]])
```

```python
print(info.keys())
print(info['./noTranspose'].keys())
print(info['./noTranspose']['32'].keys())
```

```python
times = {}
for key in info['./noTranspose'].keys():
    times[key] = []
    for k in info['./noTranspose'][key].keys():
        times[key].append(info['./noTranspose'][key][k])
```

```python
print(times)
```

```python
plt.figure(num=None, figsize=(8, 4.5), dpi=80, facecolor='w', edgecolor='k')
plt.ylabel("Time (s)")
plt.xlabel("Number of rows/columns")
for key in times:
    plt.plot(times[key],label=key)
plt.legend()
plt.yscale('log')
plt.xticks(range(len(times['1'])), [2**(i+1) for i in range(len(times['1']))])
plt.show()
```
